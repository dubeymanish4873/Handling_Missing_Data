### If we have to replace the null value of the data set then:-)
### for numerical data type we use mean/edian value of that column (Generally we prefer median).
### for categorical  data type we use mode value.

```python
## Let's first we use age one:-)
mean_age=df["age"].mean()
mean_age
```

```python
median_age=df["age"].median()
median_age
```

```python
## we use fillna command to replace the null one with mean_value
df["age"].fillna(mean_age,inplace=True)
```

```python
df.isnull().sum()
```

```python
## now we can see that all the null value is replace by the mean_value.
df.head(10)
```

```python
## Intially the 5th row age column is null but now it has been replaced by the mean value.
```

```python
## Now lets use the categical data. Eg->deck or embark_town
mode_deck=df["deck"].value_counts(dropna=True).idxmax()
mode_deck
```

```python
df["deck"].fillna(mode_deck,inplace=True)
```

```python
df.head(10)
```

```python
## As we can see we have successfully replace all the null value of deck column.
```

```python
df.isnull().sum()
```

```python
embarked_mode=df["embarked"].value_counts(dropna=True).idxmax()
embarked_mode
```

```python
df["embarked"].fillna(embarked_mode,inplace=True)
```
